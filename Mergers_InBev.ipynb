```python
import urllib.request, urllib.parse, urllib.error
from bs4 import BeautifulSoup
import re
html = urllib.request.urlopen('https://en.wikipedia.org/wiki/AB_InBev_brands').read()
soup = BeautifulSoup(html, 'html.parser')
b_ = re.findall('<li>(.*)<', soup.decode())
#print(b_.index('ZiegenBock</li>'))
b_global = b_[:26]
#print(b_.index('10 Barrel Brewing Co (USA)'))
#print(b_.index('Brutal fruit'))
b_americas = b_[227:387]
b_us = list(filter(lambda x: 'USA' in x or 'Michelob' in x or 'Bud' in x or 'Kokanee' in x or 'Lakeport' in x, b_americas))
def process(ls):
    for i in range(len(ls)):
        if ls[i][0] == '<':
            ls[i] = ls[i].split('>')[1]
    ls = map(lambda x: x.split(' (')[0].split('<')[0].replace('\\', ''), ls)
    return list(ls)

b_ = process(b_global) + process(b_us)
b_.extend(('Babe Wines', 'Best Damn Brewing Company', 'Best Damn Root Beer', 'Best Damn Apple Ale',
           'Devils Backbone Brewing Company', 'Steeler', 'Brava', 'Wee Willy'))
brands_i = list(set(b_))
brands_i
```
